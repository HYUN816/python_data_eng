```python
import nltk
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize, RegexpTokenizer
```

```python
article = open('article.txt',encoding='utf-8')
article = article.read()
```

```python
sent_tokenize(article[:200])
```

```python
word_tokenize(article[:200])[0:4]
```

```python
retokenize = RegexpTokenizer("[\w]+")
retokenize.tokenize(article[:500])[10:20]
```

```python
from nltk import Text
```

```python
text = Text(retokenize.tokenize(article))
```

```python
text.plot(20)
```

```python
from nltk import FreqDist
from nltk.tag import pos_tag
```

```python
stopwords = ["you", "of", "are", "am", "I", "a","it","by","and","from","s","i","t"]
article_tokens = pos_tag(retokenize.tokenize(article))
names_list = [t[0] for t in article_tokens if t[1] == "NN" and t[0] not in stopwords]
fd_names = FreqDist(names_list)
```

```python
fd_names.plot(10)
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
wc=WordCloud(width=1000, height=600,background_color='white', random_state=0)
plt.imshow(wc.generate_from_frequencies(fd_names))
plt.axis('off')
plt.show()
```
