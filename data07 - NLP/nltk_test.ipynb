```python
# collections 
# list(순서), dictionary(key,value)
# tuple(순서, 읽기전용)
# set(중복미허용, 순서없음)
```

```python
stopwords = set(line.strip() for line in open('stopwords.txt')) # 익명함수, 클러저
stopwords #set 구조
```

```python
# set에서 추가하는 경우 (합집합)
# union = 비파과함수
stopwords = stopwords.union(set(['the','to','one']))
stopwords
```

```python
# 대소문자 구분 하는 언어는 case sensitive 하다 
wordcount = {}
for word in a.lower().split() : 
    word = word.replace(".","")
    word = word.replace(",","")
    word = word.replace(":","")
    word = word.replace("'\'","")
    word = word.replace("*","")
    
    if word not in stopwords : 
        if word not in wordcount :
            wordcount[word]=1
        else : 
            wordcount[word]+=1
        
    
```

```python
pos = open('posreview.txt', encoding='utf-8')
# open은 stream을 연것 -> read를 통해 읽어줘야한다
p_review = pos.read()
```

```python
p_wordcount = {}
for word in p_review.lower().split() : 
    word = word.replace(" ","")
    word = word.replace(".","")
    word = word.replace(",","")
    word = word.replace(":","")
    word = word.replace("'\'","")
    word = word.replace("*","")
    
    if word not in stopwords : 
        if word not in p_wordcount :
            p_wordcount[word]=1
        else : 
            p_wordcount[word]+=1
p_wordcount[""]=0
```

```python
p_wordcounter = collections.Counter(p_wordcount)
p_wordcounter.most_common(10)
```

```python
p_list = p_wordcounter.most_common(10)
df_p = pd.DataFrame(p_list, columns=['word','count'])
df_p
```
