```python
# 데이터 프레임에 넣어라 
# 데이터 컬럼명 
import pandas as pd
import numpy as np

df = pd.DataFrame(iris_data.data, columns=iris_data.feature_names)
df
```

```python
df['target'] = iris_data.target
df
```

```python
df.columns=['sepal length', 'sepal width', 'petal length', 'petal width', 'target']
# df.rename(columns = {'sepal length (cm)' : 'sepal length'}, inplace = True)

df
```

```python
# 공부할 데이터와 테스트용 데이터를 분리 
from sklearn.model_selection import train_test_split
```

```python
x_data = df.loc[:,'sepal length':'petal width'] # 문제데이터
y_data = df.loc[:, 'target'] #정답데이터
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    x_data, # 문제 데이터
    y_data, # 정답 데이터
    test_size =0.2, # 비율(train,test)
    shuffle = True, # 셔플링 여부 
    random_state = 20 # 시드값 
    )
```

```python
# 공부방법론을 정해서 공부를 시키자 
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors = 3)  #하이퍼파라미터(내가 조절할 수 있는 파라미터)
```

```python
knn.fit(x_train, y_train)
```

```python
# 테스트용 데이터로 점수가 얼마나 나오는지 검증해보자
pred = knn.predict(x_test)
pred
```

```python
# 모의고사 점수가 얼마가 나왔는지 확인해보자 -> 채점 
from sklearn.metrics import accuracy_score
```

```python
jumsu = accuracy_score(y_test, pred) #원래정답, 내가푼답

jumsu #accuracy 
```
