```python
from sklearn.datasets import load_wine

wine_data = load_wine()
wine_data
```

```python
keys = wine_data.keys()
keys
```

```python
print('\n feature_names 의 type:',type(wine_data.feature_names))
print(' feature_names 의 shape:',len(wine_data.feature_names))
print(wine_data.feature_names)

print('\n target_names 의 type:',type(wine_data.target_names))
print(' feature_names 의 shape:',len(wine_data.target_names))
print(wine_data.target_names)

print('\n data 의 type:',type(wine_data.data))
print(' data 의 shape:',wine_data.data.shape)
print(wine_data['data'])

print('\n target 의 type:',type(wine_data.target))
print(' target 의 shape:',wine_data.target.shape)
print(wine_data.target)
```

```python
import pandas as pd
import numpy as np

df = pd.DataFrame(wine_data.data, columns=wine_data.feature_names)
df
```

```python
df['target'] = wine_data.target
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_data = df.loc[:,'alcohol':'proline'] # 문제데이터
y_data = df.loc[:, 'target'] #정답데이터
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    x_data, # 문제 데이터
    y_data, # 정답 데이터
    test_size =0.2, # 비율(train,test)
    shuffle = True, # 셔플링 여부 
    random_state = 20 # 시드값 
    )
```

```python
x_train, x_test, y_train, y_test
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors = 7) 
```

```python
knn.fit(x_train, y_train)
```

```python
pred = knn.predict(x_test)
pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
jumsu = accuracy_score(y_test, pred)

jumsu #accuracy 
```

```python
from sklearn.svm import SVC
```

```python
svc = SVC(kernel = 'rbf')
# kernel : 'linear' : 직선, 'rbf' : 곡선
```

```python
svc.fit(x_train, y_train)
```

```python
pred2 = svc.predict(x_test)
pred2
```

```python
jumsu2 = accuracy_score(y_test, pred2)

jumsu2
```
